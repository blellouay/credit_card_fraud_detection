## 2. Analyze Class Distribution

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Class counts
class_counts = df['Class'].value_counts()

print("Class Distribution:")
print(class_counts)

# Percentages
fraud_pct = (class_counts[1] / len(df)) * 100
legit_pct = (class_counts[0] / len(df)) * 100

print(f"\nLegitimate Transactions: {legit_pct:.4f}%")
print(f"Fraudulent Transactions: {fraud_pct:.4f}%")
````

````python
percentages = df['Class'].value_counts(normalize=True).sort_index() * 100

plt.figure(figsize=(6,4))
sns.barplot(
    x=['Legit', 'Fraud'],
    y=percentages.values
)

plt.title('Class Distribution (%)')
plt.xlabel('Class')
plt.ylabel('Percentage (%)')

# Add percentage labels
for i, v in enumerate(percentages.values):
    plt.text(i, v + 0.1, f'{v:.2f}%', ha='center')

plt.show()
````

### Interpretation

The class distribution reveals a severe imbalance in the dataset. Legitimate transactions account for approximately **99.83%** of all observations, while fraudulent transactions represent only **0.17%**.

This imbalance is a common characteristic of real-world fraud detection problems, where fraudulent activities are rare compared to normal transactions. As a result, **accuracy is not an appropriate evaluation metric**, since a model that predicts every transaction as legitimate would still achieve an accuracy close to 99.83% while failing to detect any fraud.

Therefore, this project will focus on metrics that better capture fraud detection performance, such as:

- **AUPRC (Area Under the Precision-Recall Curve)** as the primary metric.
- **Precision**, to measure how many detected frauds are actually fraudulent.
- **Recall**, to measure how many fraudulent transactions are successfully detected.
- **F1-Score**, to balance precision and recall.

The severe class imbalance also suggests that special techniques such as **class weighting**, **undersampling**, **oversampling**, or **SMOTE** may be required during model training.

## 5. Transaction Amount Analysis

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df['Amount_log'] = np.log1p(df['Amount'])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Original distribution with class distinction
sns.histplot(
    data=df,
    x='Amount',
    hue='Class',
    bins=50,
    stat='density',
    common_norm=False,
    alpha=0.5,
    ax=axes[0]
)

axes[0].set_title('Original Amount Distribution by Class')
axes[0].set_xlabel('Amount')
axes[0].set_ylabel('Density')

# Log-transformed distribution with class distinction
sns.histplot(
    data=df,
    x='Amount_log',
    hue='Class',
    bins=50,
    stat='density',
    common_norm=False,
    alpha=0.5,
    ax=axes[1]
)

axes[1].set_title('Log-Transformed Amount Distribution by Class')
axes[1].set_xlabel('log(Amount + 1)')
axes[1].set_ylabel('Density')

# Rename legend labels
for ax in axes:
    legend = ax.get_legend()
    if legend:
        legend.set_title('Class')
        legend.texts[0].set_text('Legit')
        legend.texts[1].set_text('Fraud')

plt.tight_layout()
plt.show()
````

### Interpretation of Transaction Amount Distribution by Class

The left plot shows the original transaction amount distribution for legitimate and fraudulent transactions. Both distributions are highly concentrated near small transaction amounts, with a long right tail caused by a limited number of high-value transactions. Due to the presence of extreme values, the majority of observations are compressed near zero, making it difficult to identify meaningful differences between the two classes.

To improve interpretability, a logarithmic transformation was applied to the transaction amount. The right plot shows the resulting distributions on the log scale. After transformation, the data becomes much more spread out, revealing differences that were not visible in the original distribution.

Several observations can be made:

- Fraudulent transactions appear to have a different distribution shape compared to legitimate transactions.
- Fraud cases are more concentrated around specific transaction amount ranges, suggesting that transaction amount may contain useful information for distinguishing fraudulent behavior.
- Legitimate transactions exhibit a broader distribution across transaction amounts, while fraudulent transactions show stronger peaks in certain regions.
- Although there is considerable overlap between the two classes, the differences observed indicate that the transaction amount feature may contribute to fraud detection when combined with other variables.

Overall, the analysis suggests that transaction amount is a potentially informative feature. The logarithmic transformation improves visualization and may also improve the performance of machine learning models by reducing the influence of extreme outliers.

## 6. Time Analysis

````python
df['Time_Hour'] = df['Time'] / 3600
````

````python
fraud_by_hour = (
    df[df['Class'] == 1]
    .groupby(df['Time_Hour'].astype(int))
    .size()
)

plt.figure(figsize=(12,5))

fraud_by_hour.plot(kind='bar')

plt.title('Fraud Transactions by Time Interval')
plt.xlabel('Approximate Hour')
plt.ylabel('Fraud Count')

plt.show()
````

### Interpretation of Fraud Transactions Over Time

The figure illustrates the number of fraudulent transactions detected within each hourly interval of the observation period. The distribution of frauds is not uniform across time, as some intervals contain significantly more fraudulent transactions than others.

Several observations can be made:

- The highest concentration of frauds occurs around **hour 11**, where approximately **43 fraudulent transactions** were recorded.
- Other notable peaks are observed around **hours 7 and 27**, indicating periods of increased fraudulent activity.
- Some intervals contain very few fraud cases (e.g., hours 0, 1, 10, 22, and 45), suggesting lower fraud occurrence during these periods.
- Fraudulent transactions appear throughout the entire observation period rather than being restricted to a specific time window.

These fluctuations suggest the presence of **temporal variations** in fraudulent behavior. However, the fraud counts do not exhibit a clear recurring pattern or trend over time. Instead, fraud activity appears to occur in localized clusters or bursts at certain intervals.

Because the dataset's `Time` feature represents the number of seconds elapsed since the first transaction rather than actual calendar time, these observations should be interpreted cautiously. While time may provide useful supplementary information, it is unlikely to be a strong predictor of fraud when used alone.

Overall, the analysis indicates that fraudulent activity varies over time and occasionally occurs in concentrated periods, suggesting that the `Time` feature may contribute useful information when combined with other transaction characteristics in a machine learning model.

## 7. Feature Correlation Analysis

````python
df= pd.read_csv("../data/creditcard.csv")
corr_matrix = df.corr(numeric_only=True).round(3)

corr_matrix.style\
    .background_gradient(cmap='coolwarm')\
    .format("{:.3f}")\
    .set_caption("Correlation Matrix")\
    .set_properties(**{
        'text-align': 'center',
        'font-size': '10pt'
    })
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import math

features = [col for col in df.columns if col != 'Class']

n_features = len(features)
n_cols = 4
n_rows = math.ceil(n_features / n_cols)

fig, axes = plt.subplots(
    n_rows,
    n_cols,
    figsize=(20, 4 * n_rows)
)

axes = axes.flatten()

for i, feature in enumerate(features):
    sns.histplot(
        data=df,
        x=feature,
        hue='Class',
        stat='density',
        common_norm=False,
        bins=30,
        alpha=0.5,
        ax=axes[i],
        legend=False
    )

    axes[i].set_title(feature, fontsize=10)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

# Remove unused subplots
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

plt.suptitle(
    'Feature Distributions by Class',
    fontsize=18,
    y=1.02
)

plt.tight_layout()
plt.show()
````

### Feature Distribution Analysis by Class

The distributions of the features were examined separately for legitimate and fraudulent transactions in order to identify variables that may help distinguish between the two classes.

The analysis reveals that several variables exhibit noticeably different distributions between legitimate and fraudulent transactions. In particular, features such as **V4, V10, V11, V12, V14, V16, V17, and V18** show clear shifts in their distributions, indicating a strong relationship with fraudulent behavior.

Some notable observations include:

- **V11** shows a substantial positive shift for fraudulent transactions compared to legitimate ones.
- **V10, V12, V14, V16, and V17** exhibit strong negative shifts for fraudulent transactions, suggesting that these variables are among the most discriminative features in the dataset.
- **V4** displays a clear separation between the two classes, with fraudulent transactions concentrated at higher values.
- Features such as **V13, V15, V22, V24, V25, and V26** show very similar distributions for both classes, suggesting a weaker individual contribution to fraud detection.
- The **Time** feature exhibits some variation between classes but does not appear to provide strong separation on its own.
- The **Amount** feature shows considerable overlap between fraudulent and legitimate transactions, indicating that transaction amount alone is insufficient for distinguishing fraud.

Overall, the analysis demonstrates that several transformed PCA features (V1–V28) contain valuable information for fraud detection. The noticeable differences observed in features such as **V10, V11, V12, V14, V16, and V17** suggest that these variables are likely to play an important role in the predictive models developed during the next phase of the project.

These findings are consistent with the correlation analysis and indicate that fraud detection will benefit from combining multiple features rather than relying on any single variable in isolation.
